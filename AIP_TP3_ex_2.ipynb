Rate can be considered equivalent to the mututal information before and after compression between X original data and X' the distorted data.
Thus:<br>
$$R(D) = I(X, X')$$

```
import numpy as np
import matplotlib.pyplot as plt
N=1000
mean, cov = [1,10], [[1, 0], [0, 1]]

# normal distribution dimension generation
x, y = np.random.multivariate_normal(mean, cov, N).T

plt.plot(x, y, 'x')
plt.axis('equal')
plt.show()

x.shape, y.shape
# https://stackoverflow.com/questions/17619415/how-do-i-combine-two-numpy-arrays-element-wise-in-python
data = np.insert(y, np.arange(len(x)), x)
X= data.reshape((x.shape[0],2))
```

```
# https://colab.research.google.com/github/jakevdp/PythonDataScienceHandbook/blob/master/notebooks/05.11-K-Means.ipynb#scrollTo=a92cMkgJ7wlZ
# from sklearn.cluster import KMeans
# kmeans = KMeans(n_clusters=4)
# kmeans.fit(X)
# y_kmeans = kmeans.predict(X)
# plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

# centers = kmeans.cluster_centers_
# plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);

### 
# https://www.sciencedirect.com/science/article/abs/pii/S0167865599000690
from sklearn.metrics import pairwise_distances_argmin


def random_init(X, n_clusters, rseed=2):
  rng = np.random.RandomState(rseed)
  i = rng.permutation(X.shape[0])[:n_clusters]
  centers = X[i]
  return centers


def find_clusters(X, n_clusters, rseed):
    
    # 1. Randomly choose clusters
    centers = random_init(X, n_clusters, rseed)
    
    iterations = 0
    while True:
        iterations += 1
        # 2a. Assign labels based on closest center

        labels = pairwise_distances_argmin(X, centers)
        # 2b. Find new centers from means of points
        new_centers = np.array([X[labels == i].mean(0)
                                for i in range(n_clusters)])
        
        # 2c. Check for convergence
        if np.all(centers == new_centers):
            break
        centers = new_centers
    
    return iterations ,centers, labels

rseed = int(np.random.random()*100)
print("randomg seed: ", rseed)
iterations, centers, labels = find_clusters(X, 4, rseed)
plt.scatter(X[:, 0], X[:, 1], c=labels,
            s=50, cmap='viridis');
x=[c[0] for c in centers]
y=[c[1] for c in centers]
plt.scatter(x,y, color="red")
```
