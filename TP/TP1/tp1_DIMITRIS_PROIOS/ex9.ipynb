```python
import skimage
import matplotlib.image as mping
import matplotlib.pyplot as plt
import numpy as np
import cv2
from scipy import ndimage as ndi
from  skimage.measure import label 
def toGrayScale(img):
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

parts = toGrayScale(np.asarray(mping.imread("./data/parts.png")))
plt.imshow(parts, cmap="gray")
plt.show()
```

```python
edges = skimage.feature.canny(parts)
plt.imshow(edges , cmap="gray")
edges
```

```python
fill_parts= ndi.binary_fill_holes(edges)
plt.imshow(fill_parts, cmap="gray")
```

```python
labels =label(fill_parts)
labels.shape
plt.imshow(labels, cmap="gray")
```

```python
areas = [r.area for r in skimage.measure.regionprops(labels)]
# plt.imshow(areas, cmap="gray")
```
